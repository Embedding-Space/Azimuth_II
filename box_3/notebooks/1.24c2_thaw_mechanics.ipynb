# 1.24c2: Thaw Mechanics - Momentum vs Gradient

**Purpose:** Determine whether the early frost/thaw phenomenon is driven by momentum or gradients.

## The Question

In 1.24c we observed a remarkable pattern:
- t=150-300: Tokens begin to freeze (~30% frozen at peak)
- t=300-450: **Thaw** — frozen fraction drops back to ~5%
- t=550+: Final freeze (Fimbulwinter)

## Two Hypotheses

**H1: Momentum-Driven Thaw**
- Adam's momentum ($m_t$) accumulated during early training
- Gradients drop below lattice threshold, but momentum is still large
- Momentum carries tokens forward → unfreezes them
- Eventually momentum decays → final freeze

**Prediction:** Momentum norm should remain large during thaw (t=300-450) even as gradients drop

**H2: Gradient-Driven Thaw**
- Live tokens still moving → changes gradient landscape for dead tokens
- Dead tokens that froze early get new nonzero gradients
- Brief reheating before final freeze

**Prediction:** Gradient magnitude should show secondary bump during thaw window

## What We'll Plot

1. State evolution (Classical/Quantum/Frozen) with momentum and gradient overlaid
2. Momentum norm over time for dead tokens
3. Gradient norm over time for dead tokens
4. Ratio: momentum / gradient (which dominates when?)
5. Zoomed view of thaw window (t=200-600)

## Parameters

```python
# Data path
DATA_PATH = "../tensors/Flannel/1.20h_flannel_7.safetensors"
MASK_PATH = "../tensors/Flannel/live_dead_tokens.safetensors"

# Thaw window (from 1.24c observations)
THAW_START = 200
THAW_END = 600

# Plot settings
DPI = 200
FIGSIZE = (12, 6)

print("✓ Parameters set")
```

## Imports

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from safetensors.torch import load_file
from pathlib import Path

print("✓ Imports complete")
```

## Device Detection

```python
if torch.cuda.is_available():
    device = 'cuda'
elif torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'

print(f"Using device: {device}")
```

## Load Data

```python
print("Loading Flannel 7 data...\n")

data = load_file(DATA_PATH)

# Extract tensors
W = data['W'][0].float()  # (1001, 10000, 64)
grad_W = data['grad_W'][0].float()  # (1001, 10000, 64)
momentum_W = data['momentum_W'][0].float()  # (1001, 10000, 64)
variance_W = data['variance_W'][0].float()  # (1001, 10000, 64)

# Load token masks
mask_data = load_file(MASK_PATH)
dead_mask = mask_data['dead_mask'].bool()
live_mask = mask_data['live_mask'].bool()

print(f"W shape: {tuple(W.shape)}")
print(f"Dead tokens: {dead_mask.sum().item()}")
print(f"\n✓ Data loaded")
```

## Compute Norms

```python
print("Computing gradient and momentum norms...\n")

# Extract dead tokens
W_dead = W[:, dead_mask, :]  # (1001, 3699, 64)
grad_W_dead = grad_W[:, dead_mask, :]  # (1001, 3699, 64)
momentum_W_dead = momentum_W[:, dead_mask, :]  # (1001, 3699, 64)
variance_W_dead = variance_W[:, dead_mask, :]  # (1001, 3699, 64)

# Compute per-token norms
grad_norms = torch.norm(grad_W_dead, dim=2)  # (1001, 3699)
momentum_norms = torch.norm(momentum_W_dead, dim=2)  # (1001, 3699)

# Mean across tokens
mean_grad_norm = grad_norms.mean(dim=1)  # (1001,)
mean_momentum_norm = momentum_norms.mean(dim=1)  # (1001,)

# Also compute displacement (for reference)
displacements = torch.norm(W_dead[1:] - W_dead[:-1], dim=2)  # (1000, 3699)
mean_displacement = displacements.mean(dim=1)  # (1000,)

print(f"Gradient norms: {tuple(grad_norms.shape)}")
print(f"Momentum norms: {tuple(momentum_norms.shape)}")
print(f"Mean gradient norm: {tuple(mean_grad_norm.shape)}")
print(f"Mean momentum norm: {tuple(mean_momentum_norm.shape)}")
print(f"\n✓ Norms computed")
```

## Load State Classification from 1.24c

```python
# Recompute state classification (simplified from 1.24c)
print("Classifying token states...\n")

# Compute ULPs
_, exponents = torch.frexp(W_dead[:-1].abs())
ulps = 2.0 ** (exponents - 7)
delta_W = W_dead[1:] - W_dead[:-1]
normalized_delta = delta_W / (ulps + 1e-10)
rounded = torch.round(normalized_delta)
is_quantized = torch.abs(normalized_delta - rounded) < 0.1
displacement_norm = torch.norm(delta_W, dim=2)
mean_ulp = ulps.mean(dim=2)

# Classify
frozen_mask = (displacement_norm == 0)
fraction_quantized = is_quantized.float().mean(dim=2)
is_small = displacement_norm < (10.0 * mean_ulp)
quantum_mask = (~frozen_mask) & (fraction_quantized > 0.7) & is_small
classical_mask = (~frozen_mask) & (~quantum_mask)

# Fractions
fraction_frozen = frozen_mask.float().mean(dim=1)
fraction_quantum = quantum_mask.float().mean(dim=1)
fraction_classical = classical_mask.float().mean(dim=1)

print(f"✓ States classified")
```

## Overview: Full Training Run

```python
# Three-panel plot: State evolution + Momentum + Gradient
fig, axes = plt.subplots(3, 1, figsize=(FIGSIZE[0], FIGSIZE[1]*1.5), dpi=DPI, sharex=True)

# Panel 1: State evolution
ax1 = axes[0]
classical_pct = fraction_classical.numpy() * 100
quantum_pct = fraction_quantum.numpy() * 100
frozen_pct = fraction_frozen.numpy() * 100

ax1.fill_between(range(len(frozen_pct)), 0, frozen_pct, 
                color='#3498db', alpha=0.8, label='Frozen')
ax1.fill_between(range(len(frozen_pct)), frozen_pct, frozen_pct + quantum_pct, 
                color='#f39c12', alpha=0.8, label='Quantum')
ax1.fill_between(range(len(frozen_pct)), frozen_pct + quantum_pct, 100, 
                color='#2ecc71', alpha=0.8, label='Classical')
ax1.axvspan(THAW_START, THAW_END, alpha=0.1, color='red', zorder=-1)
ax1.set_ylabel('State (%)')
ax1.set_title('Token State Evolution', fontweight='bold')
ax1.legend(loc='upper right', fontsize=9)
ax1.set_ylim(0, 100)
ax1.grid(True, alpha=0.3, axis='y')

# Panel 2: Momentum norm
ax2 = axes[1]
ax2.semilogy(mean_momentum_norm.numpy(), linewidth=2, color='purple', label='Momentum norm')
ax2.axvspan(THAW_START, THAW_END, alpha=0.1, color='red', zorder=-1)
ax2.set_ylabel('Momentum Norm (log)')
ax2.set_title('Dead Token Momentum: Accumulated Inertia', fontweight='bold')
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3, which='both')

# Panel 3: Gradient norm
ax3 = axes[2]
ax3.semilogy(mean_grad_norm.numpy(), linewidth=2, color='darkred', label='Gradient norm')
ax3.axvspan(THAW_START, THAW_END, alpha=0.1, color='red', zorder=-1, label='Thaw window')
ax3.set_xlabel('Training Step')
ax3.set_ylabel('Gradient Norm (log)')
ax3.set_title('Dead Token Gradient: Instantaneous Force', fontweight='bold')
ax3.legend(loc='upper right')
ax3.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()
```

## Momentum vs Gradient: Direct Comparison

```python
# Plot both on same axes
fig, ax = plt.subplots(figsize=FIGSIZE, dpi=DPI)

ax.semilogy(mean_momentum_norm.numpy(), linewidth=2, color='purple', label='Momentum', alpha=0.8)
ax.semilogy(mean_grad_norm.numpy(), linewidth=2, color='darkred', label='Gradient', alpha=0.8)
ax.axvspan(THAW_START, THAW_END, alpha=0.15, color='orange', zorder=-1, label='Thaw window')

# Mark key events
early_frost = 300
thaw_min = 400
final_freeze = 550

ax.axvline(early_frost, color='blue', linestyle='--', linewidth=1, alpha=0.5, label='Early frost peak')
ax.axvline(thaw_min, color='green', linestyle='--', linewidth=1, alpha=0.5, label='Thaw minimum')
ax.axvline(final_freeze, color='black', linestyle='--', linewidth=1, alpha=0.5, label='Final freeze')

ax.set_xlabel('Training Step', fontsize=12)
ax.set_ylabel('Norm (log scale)', fontsize=12)
ax.set_title('Momentum vs Gradient: Which Drives The Thaw?', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()
```

## Momentum / Gradient Ratio

```python
# Compute ratio: when does momentum dominate?
ratio = mean_momentum_norm / (mean_grad_norm + 1e-10)

fig, ax = plt.subplots(figsize=FIGSIZE, dpi=DPI)

ax.semilogy(ratio.numpy(), linewidth=2, color='darkgreen')
ax.axhline(1, color='red', linestyle='--', linewidth=2, label='Equal (ratio=1)')
ax.axvspan(THAW_START, THAW_END, alpha=0.15, color='orange', zorder=-1, label='Thaw window')

ax.set_xlabel('Training Step', fontsize=12)
ax.set_ylabel('Momentum / Gradient (log scale)', fontsize=12)
ax.set_title('When Does Momentum Dominate?', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()

print(f"\nMomentum/Gradient ratio during key periods:")
print(f"  Early (t=0-100):        {ratio[:100].mean():.2f}× (gradient dominates)")
print(f"  Thaw (t={THAW_START}-{THAW_END}):      {ratio[THAW_START:THAW_END].mean():.2f}× (momentum {'dominates' if ratio[THAW_START:THAW_END].mean() > 1 else 'weaker'})")
print(f"  Final (t={final_freeze}-1000):     {ratio[final_freeze:].mean():.2f}× (momentum {'dominates' if ratio[final_freeze:].mean() > 1 else 'weaker'})")
```

## Zoomed: Thaw Window Detail

```python
# Zoom into thaw window
fig, axes = plt.subplots(3, 1, figsize=(FIGSIZE[0], FIGSIZE[1]*1.5), dpi=DPI, sharex=True)

# Panel 1: Frozen fraction (zoomed)
ax1 = axes[0]
ax1.plot(range(THAW_START, THAW_END), frozen_pct[THAW_START:THAW_END], 
         linewidth=2, color='#3498db', marker='o', markersize=3)
ax1.set_ylabel('Frozen (%)')
ax1.set_title('Frozen Fraction During Thaw Window', fontweight='bold')
ax1.grid(True, alpha=0.3)

# Panel 2: Momentum (zoomed)
ax2 = axes[1]
ax2.plot(range(THAW_START, THAW_END), mean_momentum_norm[THAW_START:THAW_END].numpy(), 
         linewidth=2, color='purple', marker='o', markersize=3)
ax2.set_ylabel('Momentum Norm')
ax2.set_title('Momentum During Thaw', fontweight='bold')
ax2.grid(True, alpha=0.3)

# Panel 3: Gradient (zoomed)
ax3 = axes[2]
ax3.plot(range(THAW_START, THAW_END), mean_grad_norm[THAW_START:THAW_END].numpy(), 
         linewidth=2, color='darkred', marker='o', markersize=3)
ax3.set_xlabel('Training Step')
ax3.set_ylabel('Gradient Norm')
ax3.set_title('Gradient During Thaw', fontweight='bold')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Statistical Analysis: Correlation During Thaw

```python
print("="*80)
print("CORRELATION ANALYSIS: THAW WINDOW")
print("="*80 + "\n")

# Extract thaw window data
frozen_thaw = fraction_frozen[THAW_START:THAW_END]
momentum_thaw = mean_momentum_norm[THAW_START:THAW_END]
gradient_thaw = mean_grad_norm[THAW_START:THAW_END]

# Compute correlations with frozen fraction
corr_momentum = torch.corrcoef(torch.stack([frozen_thaw, momentum_thaw]))[0, 1].item()
corr_gradient = torch.corrcoef(torch.stack([frozen_thaw, gradient_thaw]))[0, 1].item()

print(f"Correlation with frozen fraction during thaw (t={THAW_START}-{THAW_END}):")
print(f"  Momentum: {corr_momentum:+.3f}")
print(f"  Gradient: {corr_gradient:+.3f}")
print()

if abs(corr_momentum) > abs(corr_gradient):
    print(f"✓ Momentum has stronger correlation with freezing ({abs(corr_momentum):.3f} vs {abs(corr_gradient):.3f})")
    if corr_momentum < 0:
        print(f"  Negative correlation: higher momentum → less freezing (supports momentum-driven thaw)")
    else:
        print(f"  Positive correlation: higher momentum → more freezing (counterintuitive)")
else:
    print(f"✓ Gradient has stronger correlation with freezing ({abs(corr_gradient):.3f} vs {abs(corr_momentum):.3f})")
    if corr_gradient < 0:
        print(f"  Negative correlation: higher gradient → less freezing (supports gradient-driven thaw)")
    else:
        print(f"  Positive correlation: higher gradient → more freezing (counterintuitive)")

print("\n" + "="*80 + "\n")
```

## Summary & Verdict

```python
print("\n" + "="*80)
print("THAW MECHANICS: SUMMARY & VERDICT")
print("="*80 + "\n")

print("Observed phenomenon:")
print(f"  • Early frost: {frozen_pct[early_frost]:.1f}% frozen at t={early_frost}")
print(f"  • Thaw: drops to {frozen_pct[thaw_min]:.1f}% frozen at t={thaw_min}")
print(f"  • Final freeze: {frozen_pct[final_freeze]:.1f}% frozen at t={final_freeze}")
print()

print("Key metrics during thaw window:")
print(f"  • Mean momentum: {momentum_thaw.mean():.4e}")
print(f"  • Mean gradient: {gradient_thaw.mean():.4e}")
print(f"  • Momentum/Gradient ratio: {ratio[THAW_START:THAW_END].mean():.2f}×")
print()

# Determine verdict
momentum_dominates = ratio[THAW_START:THAW_END].mean() > 1.5
gradient_bump = gradient_thaw.max() > gradient_thaw.min() * 2  # 2× variation

print("Evidence:")
if momentum_dominates:
    print(f"  ✓ Momentum >> gradient during thaw ({ratio[THAW_START:THAW_END].mean():.1f}× larger)")
else:
    print(f"  ✗ Momentum not clearly dominant (ratio: {ratio[THAW_START:THAW_END].mean():.2f}×)")

if gradient_bump:
    print(f"  ✓ Gradient shows variation during thaw ({gradient_thaw.max()/gradient_thaw.min():.1f}× range)")
else:
    print(f"  ✗ Gradient monotonically decreasing (no secondary bump)")

print()

# Final verdict
if momentum_dominates and not gradient_bump:
    print("🎯 VERDICT: Momentum-driven thaw")
    print("   Accumulated inertia from early training carries tokens forward")
    print("   even as gradients drop below lattice threshold.")
elif gradient_bump and not momentum_dominates:
    print("🎯 VERDICT: Gradient-driven thaw")
    print("   Changes in gradient landscape cause brief reheating.")
elif momentum_dominates and gradient_bump:
    print("🎯 VERDICT: Mixed dynamics")
    print("   Both momentum and gradient contribute to thaw.")
else:
    print("🤔 VERDICT: Unclear")
    print("   Neither hypothesis clearly supported. May be stochastic/batch effects.")

print("\n" + "="*80)
```
